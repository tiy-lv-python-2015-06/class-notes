# Cleaning up data

```python
rdu_weather.replace(-9999, np.nan)[0:4]
```

```python
rdu_weather = rdu_weather.replace(-9999, np.nan)
```

```python
rdu_weather.head()
```

```python
rdu_weather[ ['TMAX', 'TMIN', 'AWND', 'PRCP', 'SNOW'] ].head(3)
```

```python
rdu_weather = rdu_weather[ ['TMAX', 'TMIN', 'AWND', 'PRCP', 'SNOW'] ]
```

```python
rdu_weather.columns
```

```python
def to_fahrenheit(temps):
    # Why does this work? Temps is not one number.
    return (temps / 10 * (9 / 5) + 32).round(1)
```

```python
rdu_weather["temp_max"] = to_fahrenheit(rdu_weather['TMAX'])
rdu_weather["temp_min"] = to_fahrenheit(rdu_weather['TMIN'])
```

```python
rdu_weather.head()
```

```python
# Wind speed is in 10ths of meters per second.
# Precipitation is in 10ths of millimeters.
# Snowfall is in millimeters.

rdu_weather["wind_speed"] = rdu_weather.pop("AWND") / 10
rdu_weather["precipitation"] = rdu_weather.pop("PRCP") / 10
rdu_weather = rdu_weather.rename(columns={"SNOW": "snowfall"})
```

```python
rdu_weather.tail()
```

#Querying data (aka finding a needle in a haystack)

```python
#Let's find out if any days were over 90 in durham. AKA Spring Vegas styel
rdu_weather[ rdu_weather['temp_max'] > 90 ]
```

```python
print("Len:", len(rdu_weather[rdu_weather['temp_max'] > 90]))
print("Shape:", rdu_weather[rdu_weather.temp_max > 90].shape)
print("Count:", rdu_weather[rdu_weather.temp_max > 90].count())
```

```python
rdu_weather.temp_max.max()
```

```python
rdu_weather.temp_min.min()
```

```python
#What are the top 10 hottest days?
rdu_weather[rdu_weather.temp_max > 90].sort("temp_max")[::-1].head(10)
```

```python
rdu_weather[rdu_weather.temp_max > 90].plot()
```

```python
#Hottest days
rdu_weather[ rdu_weather.temp_max
            == rdu_weather.temp_max.max()]
```

```python
#Wettest Day
rdu_weather[ rdu_weather.precipitation == rdu_weather.precipitation.max()]
```

#Playing with Time (series)

```python
rdu_weather.precipitation.resample('M', how=["sum", "mean"]).sort("sum")[::-1]
```

```python
rdu_weather.temp_max.resample('M', how=["mean", "median", "std"])
```

```python
rdu_weather.resample("M")
```

```python
rdu_weather.temp_max.resample('W', how=["mean", "median", "std"])
```

```python
rdu_weather[ ["temp_max", "temp_min" ]].resample('W', how=["mean", "median", "std"])
```

```python
rdu_weather.temp_max.resample('MS', how=["median", "std"]).plot()
```

```python
rdu_weather.temp_max.resample('W', how=["median", "std"]).plot()
```

```python
rdu_weather[["temp_max", "temp_min"]].resample('MS', how=["mean", "median"]).plot()
rdu_weather[["temp_max", "temp_min"]].resample('W', how=["mean", "median"]).plot()
plt.show()
```
