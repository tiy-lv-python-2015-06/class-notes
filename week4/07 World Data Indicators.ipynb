```python
# Let's grab the crime-ish columns
crimedf = df[ df['Indicator Code'].isin(['IC.FRM.BRIB.ZS', 'IC.FRM.CMPU.ZS', 'IC.FRM.CORR.ZS', 'IC.FRM.CRIM.ZS'])]
```

```python
crimedf.head()
```

```python
crimedf.info()
```

The data is showing that most countries are reported so lets take the best of the years

```python
crime_pivot=crimedf.pivot_table(index='Country Code', values='2013', columns="Indicator Code")
```

```python
crime_pivot
```

```python
crime_pivot = crime_pivot.dropna(thresh=4)
```

```python
crime_pivot.head()
```

# Correlation

```python
crime_pivot['IC.FRM.BRIB.ZS'].corr(crime_pivot['IC.FRM.CMPU.ZS'])
```

Not too strong of one

```python
crime_pivot['IC.FRM.BRIB.ZS'].corr(crime_pivot['IC.FRM.CORR.ZS'])
```

0.8 is really close to the +1 of the correlation scale.  It makes sense since the bribery and corruption would go hand in hand.

```python
crime_elect = df[ df['Indicator Code'].isin(['IC.FRM.CRIM.ZS', 'IC.FRM.OUTG.ZS'])]
```

```python
crime_elect.head()
```

```python
crime_elect = crime_elect[['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code',
            '2010', '2011', '2012', '2013', '2014']]
```

```python
crime_elect.head()
```

```python
crime_elect.info()
```

```python
ce_pivot = crime_elect.pivot_table(index="Country Code", columns="Indicator Code", values='2013')
```

```python
ce_pivot = ce_pivot.dropna(thresh=2)
```

```python
ce_pivot.head()
```

```python
ce_pivot['IC.FRM.CRIM.ZS'].corr(ce_pivot['IC.FRM.OUTG.ZS'])
```
